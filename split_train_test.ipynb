# Séparation train / test

```python
import pandas as pd

df = pd.read_csv('data/dataframe_clean.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
import pandas as pd

date_debut = pd.to_datetime("2017-02-01")
date_fin = pd.to_datetime("2017-08-31")
df_train = df[(pd.to_datetime(df['Date']) >= date_debut) & (pd.to_datetime(df['Date']) <= date_fin)]

date_debut_test = pd.to_datetime("2017-09-01")
date_fin_test = pd.to_datetime("2017-11-30")
df_test = df[(pd.to_datetime(df['Date']) >= date_debut_test) & (pd.to_datetime(df['Date']) <= date_fin_test)]
```

```python
print(df_train.shape)
df_train.head()
```

```python
print(df_test.shape)
df_test.head()
```

```python
df_train['FlagImpaye'] = pd.to_numeric(df_train['FlagImpaye'], errors='coerce')
print(df_train['FlagImpaye'].dtypes)
```

```python
df_train.to_csv('data/dataframe_train.csv')
df_test.to_csv('data/dataframe_test.csv')
```
